```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
corr = np.load('enc_layers.0.residual_extrapolate_2.regreesor.weight.npy')
plt.figure(figsize=(10, 3))
sns.heatmap(corr.T[::-1], vmax=5, vmin=0)
plt.ylabel("Backward Variables", fontdict={"fontsize": 13})
plt.xlabel("Forward Variables", fontdict={"fontsize": 13})
plt.savefig("short_term_contribution.pdf", bbox_inches='tight')
```
