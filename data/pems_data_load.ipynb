```python
import pandas as pd

# 定义要处理的文件列表
file_numbers = [3, 4, 5, 6, 7, 8, 10, 11, 12]
file_prefix = 'pems'
file_suffix = '_all_common_flow.csv'

# 遍历文件列表
for num in file_numbers:
    file_name = f'{file_prefix}{num:02d}{file_suffix}'
    # 读取CSV文件
    df = pd.read_csv(file_name)
    # 修改列名
    df = df.rename(columns={'Timestamp': 'date'})
    # 保存修改后的数据回CSV文件
    df.to_csv(file_name, index=False)

print("所有文件都已处理完毕。")
```
